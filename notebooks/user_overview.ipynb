# top 10 handsets used by the customers

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
data = pd.read_csv('../assets/data/cleaned_dataset_Week2.csv')  

# Step 2: Identify the top 10 handsets
top_handsets = data['Handset Type'].value_counts().head(10)

# Step 3: Display the results
print("Top 10 Handsets Used by Customers:")
print(top_handsets)

# Step 4: Visualize the results
plt.figure(figsize=(10, 6))
top_handsets.plot(kind='bar', color='skyblue')
plt.title('Top 10 Handsets Used by Customers', fontsize=16)
plt.xlabel('Handset', fontsize=12)
plt.ylabel('Number of Users', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Step 5: Save the results to a CSV file
top_handsets.to_csv('../assets/data/top_10_handsets.csv', index=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
data = pd.read_csv('../assets/data/cleaned_dataset_Week2.csv') 

# Step 2: Extract the manufacturer from handset names (if not in a separate column)
data['Manufacturer'] = data['Handset Type'].str.split(' ').str[0]  # Adjust as per your dataset

# Step 3: Identify the top 3 manufacturers
top_manufacturers = data['Manufacturer'].value_counts().head(3)
print("Top 3 Handset Manufacturers:")
print(top_manufacturers)

# Step 4: Identify the top 5 handsets per top manufacturer
top_handsets_per_manufacturer = {}
for manufacturer in top_manufacturers.index:
    # Filter data for the manufacturer and count handset occurrences
    top_handsets = data[data['Manufacturer'] == manufacturer]['Handset Type'].value_counts().head(5)
    top_handsets_per_manufacturer[manufacturer] = top_handsets

# Step 5: Display the results
print("\nTop 5 Handsets for Each Top Manufacturer:")
for manufacturer, handsets in top_handsets_per_manufacturer.items():
    print(f"\n{manufacturer}:")
    print(handsets)

# Step 6: Visualize the results
plt.figure(figsize=(12, 8))

for i, (manufacturer, handsets) in enumerate(top_handsets_per_manufacturer.items(), start=1):
    plt.subplot(1, 3, i)  # Create a subplot for each manufacturer
    handsets.plot(kind='bar', color='skyblue')
    plt.title(f'Top 5 Handsets - {manufacturer}', fontsize=14)
    plt.xlabel('Handset', fontsize=10)
    plt.ylabel('Number of Users', fontsize=10)
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()

# Step 7: Save the results to CSV files
for manufacturer, handsets in top_handsets_per_manufacturer.items():
    handsets.to_csv(f'top_5_handsets_{manufacturer}.csv', index=True)
```
